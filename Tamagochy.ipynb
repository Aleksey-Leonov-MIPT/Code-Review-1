```python
import random
import datetime
class Tamagochy:        
    def __init__(self,name="Pet"):
        print("Hello! You can feed me, see my status and make me happy")
        if type(name) != str:
            raise AttributeError("It's a strange name for your pet  " )       
        self.name = name
        self.born_time = datetime.datetime.now()
        self.time = datetime.datetime.now()
        self.hunger = 50
        self.mood = 50
        self.Death = False
        self.dictionary = []
        print("{name} is {Hunger} % hungry and {Mood}% happy".format(name = self.name, 
                                                                     Hunger = round(self.hunger,1),
                                                                    Mood = round(self.mood,1)))
                                                            
    def __repr__(self):
        return self.status()
    def status(self):
        self.update()
        print(self.Death_check() )
        if self.Death == False:
            return "{name} is {Hunger}% hungry and {Mood}% happy".format(name = self.name,
                                                                         Hunger = round(self.hunger,1),
                                                                         Mood = round(self.mood,1))
    def update(self):
        if self.Death == True:
            print("%s is gone" % self.name)
            #return None
        else:
            now = datetime.datetime.now()
            delta = now - self.time
            self.time = now
            self.hunger -= delta.seconds/3600
            self.mood -= delta.seconds/1800
            #return None                                                                          hunger = round(self.hunger,2),
    def Death_check(self):
        if self.hunger < 0:
            self.Death = True
            return "%s died of starvation" %self.name
        if self.hunger > 100:
            self.Death = True
            return "%s died of overeating" %self.name
        if (self.time -self.born_time).seconds/3600/24/365 >1:
            self.Death = True
            return "%s was too old" %self.name
        if self.mood < 0 :
            self.Death = True
            return "%s got upset and went away" %self.name 
        return ""
    def Make_happy(self):
        self.update()
        self.mood += 5
        return self.status()
    def feed(self):
        print(self.Death_check() )
        if self.Death ==True:
            print("%s is gone" % self.name) 
            return None
        else:
            delta = int(input("How much do you want to feed me? Be careful "))
            while delta+ self.hunger >= 100:
                print("No,it's too much,be careful")
                delta = int(input("Once again: \n How much do you want to feed me? Be careful "))
            self.hunger += delta
            self.update()        
            return self.status()
    def study(self, word):
        self.update()
        if type(word) != str:
            raise AttributeError("It's not a word!" )
        else:
            self.dictionary.append(word)
    def say(self, word):
        self.update()
        if type(word) != str:
            raise AttributeError("It can't pronounce that!" )
        else:
            if word in self.dictionary:
                print(word+"!")
            else:
                print("o_o ??? o_o")
```
